```python
import os
import pandas as pd
import numpy as np
import time
from collections import Counter
import csv

dataset_dir = '/Users/nouf/Documents/KFUPM1-2/Bioninformatics/project/raw_datasets'
all_snps_ids = []

def parse23andme(sub_dir, filein):
    #counter = 0
    print("Parsing {} ...".format(extractFileName(filein.name)))
    user_snps = []
    for line in filein:

        # skip comments
        if "#" in line:
            continue
        lineArray = line.replace(" ", "").replace("\n","").split("\t")
        # ignoring SNPs that has incorrect values
        #print(lineArray[3])
        if lineArray[3] == "--" or lineArray[3] == "A" or lineArray[3] == "G" or lineArray[3] == "T" or lineArray[3] == "C":
            #print("skip")
            continue
        all_snps_ids.append(lineArray[0])

def extractFileName(path):
    strings = path.split("/")
    filename = strings[len(strings)-1]    
    return filename

# Loop through users files
file_counter = 0
for sub_dir in os.listdir(dataset_dir):
    print(sub_dir)
    sub_dir_path = os.path.join(dataset_dir, sub_dir)
    if os.path.isdir(sub_dir_path):  
        for filename in os.listdir(sub_dir_path):
            print(file_counter)
            #if file_counter == 10:
            #break
            if "23andme" not in filename:
                print("File called " + filename + " doesn't seem to be a proper file.")
                continue
            # print(filename)
            file_counter = file_counter + 1
            print("Opening file number: {} - name: {} ...".format(file_counter, filename))
            filein = open(sub_dir_path + "/" +filename)
            parse23andme(sub_dir, filein)
            filein.close()

print("Counting occurance of each SNP...")
counter = Counter(all_snps_ids)
#print(counter)
print("Filtering shared SNP...")
new_list = [snp for snp in all_snps_ids if counter[snp] == 120]
new_list = set(new_list)
#print(Counter(new_list))
print(len(all_snps_ids))
print(len(new_list))

with open("shared_snps.csv", 'w') as myfile:
    wr = csv.writer(myfile)
    wr.writerow(["number of all SNPs: ", len(all_snps_ids)])
    wr.writerow(["number of shared SNPs: ", len(new_list)])
    wr.writerow(new_list)
print("DONE!!")
        
```
