```python
import os
import pandas as pd
import numpy as np
import time
from collections import Counter
import csv
import time

dataset_dir = '/Users/nouf/Documents/KFUPM1-2/Bioninformatics/project/raw_datasets'
data = []
final_snps = []
all_df = []

def addAlleles(sub_dir, filein):
    user_snps = []
    #counter = 0
    print("Parsing {} ...".format(extractFileName(filein.name)))
    user_snps = []
    for line in filein:
        #counter = counter + 1
        #print(counter)
        #if counter == 100:
        #break
        # skip comments
        if "#" in line:
            continue
        lineArray = line.replace(" ", "").replace("\n","").split("\t")
        user_snps.append({"id": lineArray[0], extractFileName(filein.name): lineArray[3]})
 
    return user_snps
    

def extractFileName(path):
    strings = path.split("/")
    filename = strings[len(strings)-1]    
    return filename

start_time = time.time()
with open('shared_snps.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
    data = data[2]


counter = 1
for sub_dir in os.listdir(dataset_dir):
    print(sub_dir)
    sub_dir_path = os.path.join(dataset_dir, sub_dir)
    if os.path.isdir(sub_dir_path):
        for filename in os.listdir(sub_dir_path):
            if "23andme" not in filename:
                print("File called " + filename + " doesn't seem to be a proper file.")
                continue
            print("Opening file number: {} ...".format(counter))
            filein = open(sub_dir_path + "/" +filename)
            user_snps = addAlleles(sub_dir, filein)
            #print("length of user snps")
            #print(len(user_snps))
            filein.close()
            data_set = set(data)
            final_snps = [el for el in user_snps if el["id"] in data_set]
            #print("length of final_snps")
            #print(len(final_snps))
            df = pd.DataFrame(final_snps)
            df = df.set_index("id")
            df = df.transpose()
            #print(df.head())
            df.to_csv("{}/{}.csv".format(sub_dir,filename), sep = ' ')
            all_df.append(df)
            counter = counter + 1

end_time = time.time()
print("time to execute:")
print(end_time - start_time)
```
